```python
newData = pd.read_excel('sz50.xlsx',sheetname='600036.XSHG')
close = newData.loc[:,"close"]
rolling_mean = close.rolling(window=10,center=False).mean().tail()
arr_1 = np.array(rolling_mean)
arr_1
```

```python
import matplotlib.pyplot as plt
stock1 = pd.read_excel('sz50.xlsx',sheetname='600036.XSHG',index_col='datetime').close
rolling_mean = stock1.rolling(window=10).mean()
rolling_mean.name = "10days rolling "

plt.figure(figsize=(15,7))
stock1.plot()
rolling_mean.plot()
plt.title("Stock price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()
```
